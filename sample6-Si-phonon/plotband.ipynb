```python
import numpy as np
import matplotlib.pyplot as plt

file_path = 'Si.freq.gp'
freq=np.loadtxt(file_path)

nmodes=freq.shape[1]-1
nmodes
```

```python
for mode in range(nmodes):
    plt.plot(freq[:,0],freq[:,mode+1],label='mode %d'%(mode+1))

plt.title('phonon dispersion')
plt.ylabel('frequency (cm^-1)')

plt.savefig('phonon_dispersion.png',dpi=600)
```
